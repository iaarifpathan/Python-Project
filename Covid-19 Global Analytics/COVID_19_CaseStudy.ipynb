## Group by + Aggregation

<div style="padding: 12px; background-color: #f8f9fa; border-left: 4px solid #0d6efd; border-radius: 4px;"><strong>📊 Grouping by Country & Date:</strong><br>Currently, there are multiple rows for a single country on the same date (because large countries have data split by states). We use <code>groupby()</code> and <code>sum()</code> to add all the states together. Now, we have one clean row showing the total cases per country, per day.</div>

```python
# cd stands for country and date
confirmed_cd = (
    confirmed_long.groupby(['Country/Region', 'Date'])['Confirmed'].sum().reset_index()
)
confirmed_cd
```

```python
deaths_cd = (
    deaths_long.groupby(['Country/Region', 'Date'])['Deaths'].sum().reset_index()
)
deaths_cd
```

```python
recovered_cd = (
    recovered_long.groupby(['Country/Region', 'Date'])['Recovered'].sum().reset_index()
)
recovered_cd
```

<div style="padding: 20px; background-color: #f4f7f6; border-left: 5px solid #2980b9; border-radius: 8px;">
    <h2 style="margin-top: 0; color: #2980b9;">Merging the data</h2>
    <p>To combine different datasets (like deaths, cases, and recoveries) into one master table, we use the <code>pandas.merge()</code> function.</p>
    <pre style="background-color: #272822; color: #f8f8f2; padding: 10px; border-radius: 5px;">pandas.merge(left, right, how='inner', on=None)</pre>
    <p><strong>Parameters:</strong></p>
    <ul>
        <li><strong>left / right:</strong> The two dataframes to combine.</li>
        <li><strong>how:</strong> Type of join (<code>'inner'</code>, <code>'left'</code>, <code>'outer'</code>).</li>
        <li><strong>on:</strong> The common column used to link tables.</li>
    </ul>
</div>

```python
# pandas.merge(left table, right table, how = 'inner', on = none)
merged = (confirmed_cd
          .merge(deaths_cd, on = ['Country/Region', 'Date'], how = 'inner')
          .merge(recovered_cd, on = ['Country/Region', 'Date'], how = 'inner')
          .sort_values(['Country/Region', 'Date'])
)
merged
```

<div style="padding: 12px; background-color: #f8f9fa; border-left: 4px solid #0d6efd; border-radius: 4px;"><strong>📈 Finding the Latest Cases:</strong><br>This dataset is <i>cumulative</i> (today's number already includes all past cases). Because of this, we don't need to add everything together. To find the total total cases a country has right now, we just need to look at the data on the <i>very last date</i> available.</div>

```python
latest_dates = merged.groupby('Country/Region')['Date'].max()
latest_dates
```

<div style="padding: 12px; background-color: #fdfdfe; border-left: 4px solid #28a745; border-radius: 4px;"><strong>⚙️ Checking the Dates:</strong><br>Good news: all countries have the exact same final recorded date (<code>2021-05-29</code>). Since everyone's timeline ends on the same day, we can simply filter the data using this specific date to find the final total cases for every country.</div>

```python
# New DataFrame()
# merged.copy() yaha humne hamare data ki ek shallow copy tayar ki hai taaki hamara original data kisi bhi tarah se impact na ho.
latest_totals = merged.copy()
latest_totals = latest_totals[latest_totals['Date'] == '2021-05-29']
latest_totals
```

<div style="padding: 12px; background-color: #fff8e1; border-left: 4px solid #ffc107; border-radius: 4px;">
    <strong>⚠️ Handling Different End Dates (Dynamic Filtering):</strong><br>
    <i>What if countries stopped reporting data on different days?</i> A single static date filter wouldn't work. Here is the fallback logic for that scenario:<br>
    <ul>
        <li><strong>1. Merge:</strong> We calculate the specific maximum date for each country and merge it back into our main dataset.</li>
        <li><strong>2. Rename:</strong> We rename this newly merged column (e.g., to <code>Latest_Date</code>) so Python doesn't get confused between the two date columns.</li>
        <li><strong>3. Query:</strong> We use the <code>.query()</code> function to compare the two columns. It checks every row and keeps only the data where the actual <code>Date</code> perfectly matches that country's <code>Latest_Date</code>.</li>
    </ul>
</div>

```python
# Merging with respect to (w.r.t) Hypothetical situation where date is not same for all countries.
latest_totals = (
    merged.merge(latest_dates.rename("LatestDate"), on = ['Country/Region'], how = 'inner').query('Date == LatestDate')
)
latest_totals
```

```python
latest_totals = latest_totals.set_index("Country/Region")['Confirmed']
latest_totals
```

```python
# Top-5 countries that has the most confirm cases.
top_5_countries = latest_totals.nlargest(5)
top_5_countries
```

<div style="padding: 12px; background-color: #f8f9fa; border-left: 4px solid #0d6efd; border-radius: 4px;"><strong>📊 Getting Ready for the Line Chart:</strong><br>To plot the line chart for the Top 5 countries, we just need their names right now. So, we are extracting just the country names from our previous result.</div>

```python
top_5_countries = latest_totals.nlargest(5).index
top_5_countries # The result is not in better format so now we convert it into better format.
```

```python
# using index.tolist() for better format.
top_5_countries = latest_totals.nlargest(5).index.tolist()
top_5_countries
```

<div style="padding: 12px; background-color: #fdfdfe; border-left: 4px solid #28a745; border-radius: 4px;"><strong>📈 Filtering the Top 5:</strong><br>Now that we have the names of the Top 5 most affected countries, we are going back to our main dataset to filter out all the daily records from start to finish, but <i>only</i> for these 5 specific countries.</div>

```python
for country in top_5_countries:
    top_countries = (merged[merged['Country/Region'] == country]).sort_values('Date')
top_countries
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a professional look
sns.set_style("whitegrid")

plt.figure(figsize=(10, 5))

# Loop through and plot
for country in top_5_countries:
    top_countries  = merged[merged['Country/Region'] == country].sort_values('Date')
    plt.plot(top_countries ['Date'], top_countries ['Confirmed'], label=country, linewidth=2)

# Professional Title and Labels
plt.title('Confirmed COVID-19 Cases in the Top 5 Affected Nations', 
          fontsize=16, fontweight='bold', color='#1E3A8A', pad=20)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Cumulative Confirmed Cases', fontsize=12)

# Improve x-axis readability
plt.xticks(rotation=45)
plt.legend(title="Country")

# Remove the top and right spines for a clean "Report" look
sns.despine()

plt.tight_layout()
plt.show()
```

<div style="padding: 12px; background-color: #f8f9fa; border-left: 4px solid #0d6efd; border-radius: 4px;">
    <strong>🎯 Targeted Filtering & Aggregation:</strong><br>
    To isolate the exact data we needed, we first defined a list of our specific target countries. We then filtered the main dataset by matching the <code>Country/Region</code> column against this list. Finally, we grouped the filtered data and sorted the results for a clean, organized output.
</div>

```python
target_countries = ['Germany','France','Italy']
country_data_total = (
    merged.loc[merged['Country/Region'].isin(target_countries)]
    .groupby(['Country/Region', 'Date'])['Confirmed'].sum().reset_index()
    .sort_values(['Country/Region', 'Date'])
)
country_data_total
```

<div style="padding: 12px; background-color: #f8f9fa; border-left: 4px solid #0d6efd; border-radius: 4px;">
    <strong>🛠️ Calculating Daily New Cases:</strong><br>
    Since the dataset is cumulative, we use the <code>.diff()</code> function to convert total counts into daily increments by subtracting the previous day's value from today's. Because the first record for each country has no 'previous day' to compare against, <code>.diff()</code> naturally creates <code>NaN</code> values. We use <code>.fillna(0)</code> to replace these with 0, ensuring our data remains clean and numerically consistent for analysis.
</div>

```python
# DailyNew_cases is a new column we create
country_data_total['DailyNew_cases'] = (
     country_data_total.groupby('Country/Region')['Confirmed'].diff()
    .fillna(country_data_total['Confirmed'])
)
country_data_total
```

```python
daily_new_matrix = (
    country_data_total.pivot(
        index = 'Date',
        columns = 'Country/Region',
        values = 'DailyNew_cases'
    ).sort_index()
)
daily_new_matrix
```

```python
daily_new_matrix['France'].nlargest(5)
```

```python
daily_new_matrix['Germany'].nlargest(5)
```

```python
daily_new_matrix['Italy'].nlargest(5)
```

```python
# target_countries = ['Germany','France','Italy']
daily_new_by_country = {}
for country in target_countries:
    if country in daily_new_matrix.columns:
        series = daily_new_matrix[country]
        daily_new_by_country[country] = series
daily_new_by_country
```

```python
# peaks {country} : (max_value, date) per date.
peaks = {}
for country, series in daily_new_by_country.items():
    max_value = int(series.max())
    max_index = series.idxmax()
    peaks[country] = (max_value, max_index)
peaks
```

```python
# Set professional theme
sns.set_style("whitegrid")

plt.figure(figsize=(10, 5))

# Plotting with professional settings
for country, series in daily_new_by_country.items():
    plt.plot(series.index, series.values, label=country, linewidth=2, alpha=0.8)

# Professional Title and Labels
plt.title('Daily Transmission Velocity: Comparative Analysis (Germany, France, Italy)', 
          fontsize=16, fontweight='bold', color='#1E3A8A', pad=20)
plt.xlabel('Date', fontsize=12)
plt.ylabel('New Daily Cases', fontsize=12)

# Styling
plt.legend(title="Country", loc='upper left')
plt.xticks(rotation=45)
sns.despine() # Removes unnecessary top/right borders
plt.tight_layout()

plt.show()
```

<div style="background-color: #f0f7ff; padding: 20px; border-left: 6px solid #1E3A8A; border-radius: 8px; font-family: sans-serif;">
    <h3 style="margin: 0 0 10px 0; color: #1E3A8A;">Problem Statement: 5.2 Compare the recovery rates between Canada and Australia as of December 31, 2020</h3>
    <p style="margin: 0; color: #333; line-height: 1.6; font-size: 14px;">
        To evaluate pandemic management efficiency, we calculate the recovery rate for Canada and Australia as of December 31, 2020. By computing the ratio of <code>Recoveries / Confirmed Cases</code>, we normalize the data to understand the proportion of patients who recovered relative to the total number of infections, allowing for a fair comparison of health outcomes between the two nations.
    </p>
</div>

```python
# .loc[rows, cols] loc mai phla filter rows par lagta hai and dusra filter columns par lagta hai.
countries = ['Canada', 'Australia']
target_date = pd.Timestamp('2020-12-31')

recovery_rates = (
    merged.loc[(merged['Country/Region'].isin(countries)) & (merged['Date'] == target_date), #Rows
        ['Country/Region','Recovered', 'Confirmed'] #Columns
    ].groupby('Country/Region')
    .sum().reset_index()
)
recovery_rates
```

```python
recovery_rates['RecoveryRate%'] = (recovery_rates['Recovered'] / recovery_rates['Confirmed']*100).round(2)
recovery_rates
```

```python
sns.set_style("whitegrid")

plt.figure(figsize=(7, 5))

# Use professional, consistent colors (Deep Navy and a softer Blue)
colors = ['#1E3A8A', '#4A90E2']

bars = plt.bar(recovery_rates['Country/Region'], 
               recovery_rates['RecoveryRate%'], 
               color=colors, 
               width=0.6)

plt.title('Comparative Recovery Efficiency: Australia vs. Canada', 
          fontsize=14, fontweight='bold', color='#1E3A8A', pad=15)
plt.ylabel('Recovery Rate (%)', fontsize=12)
plt.ylim(0, 105) # Added a small buffer to make labels look less crowded

# Adding labels clearly
plt.bar_label(bars, fmt='%.1f%%', padding=3, fontsize=11, fontweight='bold')

sns.despine() # Clean look
plt.tight_layout()
plt.show()
```

    <h3 style="margin: 0 0 10px 0; color: #1E3A8A;">Problem Statement: 7.2 Analyze the monthly sum of confirmed cases, deaths, and recoveries</h3>
    <p style="margin: 0; color: #333; line-height: 1.6; font-size: 14px;">
        To identify broad patterns in the pandemic's progression, we transition from daily data to monthly aggregates. By grouping the dataset by month, we effectively smooth out volatile daily fluctuations and uncover clearer high-level trends in infection, recovery, and mortality rates. This monthly view provides a more structured perspective for analyzing how the viral impact evolved over different phases of the year.
    </p>
</div>

```python
m = merged.copy()
m = m.sort_values(['Country/Region', 'Date'])
m['YearMonth'] = m['Date'].dt.to_period('M')
m
```

```python
#Daily Cases of Confirmed, Deaths, and Recovered
# .clip(lower = 0) ye isliye use kiya hai taaki value negative na jaye islilye numne lower bound ki value ko 0 rakha hai.
confirmed_diff = m.groupby('Country/Region')['Confirmed'].diff().clip(lower = 0)
m['DailyConfirmed'] = confirmed_diff.fillna(0)

deaths_diff = m.groupby('Country/Region')['Deaths'].diff().fillna(0).clip(lower = 0)
m['DailyDeaths'] = deaths_diff.fillna(0)

recovered_diff = m.groupby('Country/Region')['Recovered'].diff().fillna(0).clip(lower = 0)
m['DailyRecovered'] = recovered_diff.fillna(0)

m
```

```python
monthly_sum_trends = (
    m.groupby(['Country/Region', 'YearMonth'])[['DailyConfirmed', 'DailyDeaths', 'DailyRecovered']]
    .sum()
    .rename(
        columns = {
            'DailyConfirmed' : 'Confirmed',
            'DailyDeaths' : 'Deaths',
            'DailyRecovered' : 'Recovered'
        }
    ).astype(int)
    .reset_index()
    .sort_values(['Country/Region', 'YearMonth'])
)
monthly_sum_trends
```
